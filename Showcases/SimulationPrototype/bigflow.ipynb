```python
from BigFlowPrototypes import Image, Linear, Linear_Embeded

def main():
    # Ask if the data is 'image' or 'linear'
    data_type = input("Is your data 'image' or 'linear'? ").strip().lower()
    
    if data_type not in ['image', 'linear']:
        print("Invalid data type. Please enter 'image' or 'linear'.")
        return
    
    # For image data, prompt for database details
    if data_type == 'image':
        db_url = input("Enter the database URL: ")
        table_name = input("Enter the table name: ")
        dataset_name = input("Enter the dataset name: ")
        
        print("Running Image data processing...")
        return Image.main(db_url, table_name, dataset_name)
        
    elif data_type == 'linear':
        # For linear data, prompt for file and label details
        dataset_name = input("Enter the dataset name (file path): ")
        label = input("Enter the label column name: ")
        
        # Check if there is text data to embed
        has_text = input("Do you have text data that might predict the target value? (yes/no) ").strip().lower() == 'yes'
        
        if has_text:
            text_columns_input = input("Enter the text column names (comma-separated): ")
            text_columns = [col.strip() for col in text_columns_input.split(',')]
            
            print("Running Linear data processing with text embeddings...")
            return Linear_Embeded.main(dataset_name, label, text_columns)
        else:
            print("Running Linear data processing...")
            return Linear.main(dataset_name, label)


```

```python
X_train, X_test, y_train, y_test = main()
```
